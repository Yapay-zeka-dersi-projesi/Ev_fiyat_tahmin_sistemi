```python
import matplotlib.pyplot as plt

# 1. Eksik değerleri hesapla
eksik_degerler = data.isnull().sum()

# 2. Sadece eksik değeri 0'dan büyük olan sütunları al
eksik_olanlar = eksik_degerler[eksik_degerler > 0]

# Eğer hiç eksik değer yoksa uyarı ver, varsa grafiği çiz
if eksik_olanlar.empty:
    print("Veri setinde hiç eksik (null) değer bulunamadı.")
else:
    # Pasta grafiği oluşturma
    plt.figure(figsize=(10, 8), facecolor='white')
    plt.pie(eksik_olanlar, 
            labels=eksik_olanlar.index, 
            autopct='%1.1f%%', 
            startangle=140, 
            colors=plt.cm.Paired.colors)
    
    plt.title('Eksik Değerlerin Sütunlara Göre Dağılımı')
    plt.axis('equal')  # Grafiğin tam daire görünmesini sağlar
    plt.show()
```
